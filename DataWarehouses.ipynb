# Uniones - 1

```python
mergeVxPV = pd.merge(tableVendor[['businessentityid', 'name']], tableProductVendor[['productid', 'businessentityid']], 
                     how='inner', on='businessentityid')
mergeVxPV = mergeVxPV.rename(columns={'name':'name_vendor'})
```

```python
mergeVxPVxP = pd.merge(mergeVxPV, tableProduct[['productid', 'productsubcategoryid']], how='inner', on='productid')
```

```python
mergeVxPVxPxPSC = pd.merge(mergeVxPVxP, tableProductSubcategory[['productsubcategoryid', 'productcategoryid', 'name']], 
                           how='inner', on='productsubcategoryid')
mergeVxPVxPxPSC = mergeVxPVxPxPSC.rename(columns={'name':'name_productsubcategory'})
```

```python
mergeVxPVxPxPSCxPC = pd.merge(mergeVxPVxPxPSC, tableProductCategory[['productcategoryid', 'name']], 
                              how='inner', on='productcategoryid')
mergeVxPVxPxPSCxPC = mergeVxPVxPxPSCxPC.rename(columns={'name':'name_productcategory'})
```

# Loading data to BigQuery

```python
credentials = service_account.Credentials.from_service_account_file( 
    'C:/Users/aleja/Desktop/UEB V/Base de Datos 2/CORTE 3/data-warehouses-y-elts-315022-35343d2008fb.json', 
    scopes = [ 'https://www.googleapis.com/auth/cloud-platform' ]
)
```

```python
client = bigquery.Client( credentials = credentials, project = credentials.project_id )
```

```python
job_config = bigquery.LoadJobConfig(
    schema = [
        # Supported datatypes: https://cloud.google.com/bigquery/docs/reference/standard-sql/data-types
        bigquery.SchemaField( 'businessentityid', bigquery.enums.SqlTypeNames.INT64 ),
        bigquery.SchemaField( 'name_vendor', bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField( 'productid', bigquery.enums.SqlTypeNames.INT64),
        bigquery.SchemaField( 'productsubcategoryid', bigquery.enums.SqlTypeNames.FLOAT64 ),
        bigquery.SchemaField( 'productcategoryid', bigquery.enums.SqlTypeNames.INT64 ),
        bigquery.SchemaField( 'name_productsubcategory', bigquery.enums.SqlTypeNames.STRING ),
        bigquery.SchemaField( 'name_productcategory', bigquery.enums.SqlTypeNames.STRING ),
    ],
    # Drod and re-create table, if exist
    write_disposition = 'WRITE_TRUNCATE',
)
```

```python
job_config = bigquery.LoadJobConfig(
    schema = [
        # Supported datatypes: https://cloud.google.com/bigquery/docs/reference/standard-sql/data-types
        bigquery.SchemaField( 'specialofferid_x', bigquery.enums.SqlTypeNames.INT64 ),
        bigquery.SchemaField( 'type_specialofferid', bigquery.enums.SqlTypeNames.STRING),
        bigquery.SchemaField( 'productid_x', bigquery.enums.SqlTypeNames.INT64),
        bigquery.SchemaField( 'productid_specialofferid', bigquery.enums.SqlTypeNames.STRING ),
        bigquery.SchemaField( 'productid_y', bigquery.enums.SqlTypeNames.INT64 ),
        bigquery.SchemaField( 'specialofferid_y', bigquery.enums.SqlTypeNames.INT64 ),
        bigquery.SchemaField( 'orderqty', bigquery.enums.SqlTypeNames.INT64 ),
    ],
    # Drod and re-create table, if exist
    write_disposition = 'WRITE_TRUNCATE',
)
```

```python
job = client.load_table_from_dataframe(
    mergeVxPVxPxPSCxPC, 'ventasPorProducto.mergeVxPVxPxPSCxPC', job_config = job_config
)
job.result()
```

```python
job = client.load_table_from_dataframe(
    mergeSOxSOPxSOD, 'ventasEspecales.mergeSOxSOPxSOD', job_config = job_config
)
job.result()
```
