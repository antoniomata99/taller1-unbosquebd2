# Uniones - 1

```python
mergeVxPV = pd.merge(tableVendor[['businessentityid', 'name']], tableProductVendor[['productid', 'businessentityid']], 
                     how='inner', on='businessentityid')
mergeVxPV = mergeVxPV.rename(columns={'name':'name_vendor'})
```

```python
mergeVxPVxP = pd.merge(mergeVxPV, tableProduct[['productid', 'productsubcategoryid']], how='inner', on='productid')
```

```python
mergeVxPVxPxPSC = pd.merge(mergeVxPVxP, tableProductSubcategory[['productsubcategoryid', 'productcategoryid', 'name']], 
                           how='inner', on='productsubcategoryid')
mergeVxPVxPxPSC = mergeVxPVxPxPSC.rename(columns={'name':'name_productsubcategory'})
```

```python
mergeVxPVxPxPSCxPC = pd.merge(mergeVxPVxPxPSC, tableProductCategory[['productcategoryid', 'name']], 
                              how='inner', on='productcategoryid')
mergeVxPVxPxPSCxPC = mergeVxPVxPxPSCxPC.rename(columns={'name':'name_productcategory'})
```

```python
mergeVxPVxPxPSCxPC.head()
```

# Transforming data - 1

```python
grouping_1 = mergeVxPVxPxPSCxPC.groupby( ['name_vendor', 'name_productcategory'] ).agg( { 'productid': 'count' } ).reset_index().sort_values( by = 'name_vendor' )
```

```python
pivotProductosOfrecidos = grouping_1.pivot(index='name_vendor', columns='name_productcategory', values='productid')
```

# Uniones - 2

```python
mergeSOxSOP = pd.merge(tableSpecialoffer[['specialofferid', 'type']], tableSpecialofferproduct[['specialofferid', 'productid']], 
                     how='inner', on='specialofferid')
```

```python
mergeSOxSOP = mergeSOxSOP.rename(columns={'type':'type_specialofferid'})
mergeSOxSOP['productid_specialofferid'] = mergeSOxSOP['productid'].map(str) + "_" + mergeSOxSOP['specialofferid'].map(str)
```

```python
tableSalesorderdetail['productid_specialofferid'] = tableSalesorderdetail['productid'].map(str) + "_" + tableSalesorderdetail['specialofferid'].map(str)
```

```python
mergeSOxSOPxSOD = pd.merge(mergeSOxSOP, tableSalesorderdetail[['productid', 'specialofferid', 'orderqty', 'productid_specialofferid']], 
                     how='inner', on='productid_specialofferid')
```

# Transforming data - 2

```python
grouping_2 = mergeSOxSOPxSOD.groupby( ['type_specialofferid'] ).agg( { 'orderqty': 'sum' } ).reset_index().sort_values( by = 'type_specialofferid' )
```
